```python
idx = 3
# Get the lyrics from the dataset
title = lyric_head.iloc[idx]["title"]
tag = lyric_head.iloc[idx]["tag"]
lyrics = lyric_head.iloc[idx]["lyrics"]

# Perform preprocessing on the lyrics
lyrics = re.sub(
    r"\[Produced .*?\]", "", lyrics
)  # remove tags like [Produced by]

lyrics = re.sub(
    r":.*?\]", ":]", lyrics
)  # remove any possible artist names after "Chorus: " or "Verse: " etc.

lyrics = re.sub(
    r"\n{3,}", "\n\n", lyrics
)  # replace multiple newlines with two newlines

lyrics = re.sub(
    r'^\n+', '', lyrics
)

lyrics
```
